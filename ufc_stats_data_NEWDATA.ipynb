```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import time
import datetime
```

```python
results = requests.get("http://ufcstats.com/statistics/events/completed?page=all")

print(results.status_code)
```

```python
soup = bs(results.content, 'html.parser')

events = soup.find_all("a", {"class": "b-link_style_black"})

events
```

```python
eventlinks = []
for a in soup.find_all("a", {"class": "b-link_style_black"}):
    eventlinks.append(a['href'])
eventdates = []
for a in soup.find_all("span", {"class": "b-statistics__date"}):
    eventdates.append(a.text.strip())
eventdates = eventdates[1:]
eventlocs = []
for a in soup.find_all("td", {"class": "b-statistics__table-col_style_big-top-padding"}):
    eventlocs.append(a.text.strip())
eventlocs = eventlocs[1:]
```

```python
newUFCfighterlinks = []
newUFCfighternames = []
for i in eventlinks[0:2]:
    driver.get(i)
    for a in driver.find_elements_by_tag_name('a.b-link.b-link_style_black'):
        newUFCfighterlinks.append(a.get_attribute("href"))
        newUFCfighternames.append(a.get_attribute("text").strip())
    
newUFCfighters = dict(zip(newUFCfighternames,newUFCfighterlinks))
```

```python
newUFCnames = []
for i in newUFCfnames:
    newUFCnames.append(i.replace(' ', '-'))
newUFCnames
```

```python
j = []

for i in newUFCflinks[0:]:
    driver.get(i)
    time.sleep(1.2)
    for a in driver.find_elements_by_css_selector('body > section > div > div > div.b-list__info-box.b-list__info-box_style_small-width.js-guide > ul'):
        j.append(a.text.strip())
        
j
```

```python
newfbios = {}

for i in newUFCflinks[0:]:
    name = []
    info = []
    driver.get(i)
    time.sleep(1.2)
    for a in driver.find_elements_by_css_selector('body > section > div > h2 > span.b-content__title-highlight'):
        name.append(a.text.strip())
    for b in driver.find_elements_by_css_selector('body > section > div > div > div.b-list__info-box.b-list__info-box_style_small-width.js-guide > ul'):
        info.append(b.text.strip())
        
    newfbios.update(dict(zip(name,info)))

newfbios
```

```python
len(list(newfbios.values()))
```

```python
H = []
W = []
R = []
S = []
B = []

for i in list(newfbios.values()):
    H.append(i.split('\n', 4)[0])
    W.append(i.split('\n', 4)[1])
    R.append(i.split('\n', 4)[2])
    S.append(i.split('\n', 4)[3])
    B.append(i.split('\n', 4)[4])
```

```python
height = []
for i in H:
    height.append(i[8]+"-"+i[11:-1])
height
```

```python
weight = []
for i in W:
    weight.append(i[8:11])
weight
```

```python
reach = []
for i in R:
    reach.append(i[7:-1])
reach
```

```python
stance = []
for i in S:
    stance.append(i[8:])
stance
```

```python
bdays = []
for i in B:
    bdays.append(i[5:])
bdays
```

```python
newfighterbios = pd.DataFrame()
newfighterbios['fighter'] = newfnames
newfighterbios['birthday'] = bdays
newfighterbios['height'] = height
newfighterbios['weight'] = weight
newfighterbios['reach'] = reach
newfighterbios['stance'] = stance

newfighterbios
```

```python
feet = []
inch = []
for i in height:
    feet.append(i.split('-', 1)[0])
    inch.append(i.split('-', 1)[1])
```

```python
feets = []
inches = []
for i in feet:
    feets.append(pd.to_numeric(i, errors='coerce'))
for i in inch:
    inches.append(pd.to_numeric(i, errors='coerce'))
```

```python
heights = []
for i in range(len(feets)):
    heights.append(feets[i]*12 + inches[i])
```
