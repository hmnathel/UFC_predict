```python
newfightinfos = pd.read_csv('newfights.csv')
newfightinfos.drop(['Unnamed: 0'], axis=1, inplace=True)
newfightinfos
```

```python
newwinners = []
newlosers = []
for a in newfightinfos['Fighter']:
    newwinners.append(a.split('  ',1)[0])
    newlosers.append(a.split('  ',1)[1])
```

```python
newfightinfos['winner'] = newwinners
newfightinfos['loser'] = newlosers
newfightinfos.drop(['Fighter'], axis=1, inplace=True)
newfightinfos
```

```python
fightinfos = newfightinfos.append(fightinfos, ignore_index=True)
```
