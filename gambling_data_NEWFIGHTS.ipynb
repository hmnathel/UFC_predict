```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import re
import datetime
```

```python
def getLinks(url):
    results = requests.get(url)
    soup = bs(results.content, 'html.parser')
    links = []
    for link in soup.find_all('a', attrs={'href': re.compile("^mma_event_betting_history")}):
        links.append(link.get('href'))
    pre = 'https://www.betmma.tips/'
    nlinks = [pre + x for x in links]
    return nlinks
```

```python
fwinners = []
for i in range(len(newODDS)):
    if newODDS['F1'][i]=='WON':
        fwinners.append(newODDS['f1'][i])
    elif newODDS['F2'][i]=='WON':
        fwinners.append(newODDS['f2'][i])
        
fwinners
```

```python
flosers = []
for i in range(len(newODDS)):
    if newODDS['F1'][i]=='LOST':
        flosers.append(newODDS['f1'][i])
    elif newODDS['F2'][i]=='LOST':
        flosers.append(newODDS['f2'][i])
        
flosers
```

```python
fwinner_odds = []
for i in range(len(newODDS)):
    if newODDS['F1'][i]=='WON':
        fwinner_odds.append(newODDS['f1_odds'][i])
    elif newODDS['F2'][i]=='WON':
        fwinner_odds.append(newODDS['f2_odds'][i])
        
fwinner_odds
```

```python
floser_odds = []
for i in range(len(newODDS)):
    if newODDS['F1'][i]=='LOST':
        floser_odds.append(newODDS['f1_odds'][i])
    elif newODDS['F2'][i]=='LOST':
        floser_odds.append(newODDS['f2_odds'][i])
        
floser_odds
```
