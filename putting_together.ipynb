```python
fightinfos = pd.read_csv('fightinfos.csv')
fightinfos
```

```python
fight_stats_infos = pd.concat([fightinfos, fights], axis=1)
fight_stats_infos
```

```python
fight_stats_infos.drop(['Unnamed: 0'], axis=1, inplace=True)
fight_stats_infos.drop(['Fighter'], axis=1, inplace=True)
fight_stats_infos
```

```python
mins = []
secs = []
for i in range(len(fight_stats_infos)):
    mins.append(float(fight_stats_infos['Time'][i].split(':',1)[0]))
for i in range(len(fight_stats_infos)):
    secs.append(float(fight_stats_infos['Time'][i].split(':',1)[1]))
```

```python
mins
```

```python
secs
```

```python
seconds = []
for i in secs:
    seconds.append(i/60)
seconds
```

```python
minutes = []
for i,e in zip(mins,seconds):
    minutes.append(i+e)
minutes
```

```python
roundmins = []
for i in range(len(fight_stats_infos)):
    roundmins.append(float(fight_stats_infos['Round'][i] - 1))
roundmins
```

```python
tot_minutes = []
for i,e in zip(minutes,roundmins):
    tot_minutes.append(i+e)
tot_minutes
```

```python
fight_stats_infos['tot_mins'] = tot_minutes
fight_stats_infos
```
