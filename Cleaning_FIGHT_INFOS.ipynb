```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import datetime
import csv
import string
```

```python
fightinfos = pd.read_csv('fights.csv')
```

```python
newfightinfos = pd.read_csv('newfights.csv')
```

```python
fightinfos = newfightinfos.append(fightinfos, ignore_index=True)
```

```python
fightinfos['date'] = pd.to_datetime(fightinfos['date'])
fightinfos
```

```python
winners = []
losers = []
for a in fightinfos['Fighter']:
    winners.append(a.split('  ',1)[0])
    losers.append(a.split('  ',1)[1])
```

```python
fightinfos['winner'] = winners
fightinfos['loser'] = losers
```

```python
fightinfos.drop('Unnamed: 0', axis=1, inplace=True)
fightinfos.drop('Fighter', axis=1, inplace=True)
```
