```python
def get_cumul_combined_df(winner, loser, date):
    
    winnerdf = get_cumul_fighter_fight(winner, date) 
    winnerdf = winnerdf.add_suffix('_w')
    winner_new_names = {'winner_w': 'winner', 'loser_w': 'loser', 'date_w': 'date'}
    winnerdf.rename(columns = winner_new_names, inplace=True)
    
    loserdf = get_cumul_fighter_fight(loser, date)
    loserdf = loserdf.add_suffix('_l')
    loser_new_names = {'winner_l': 'winner', 'loser_l': 'loser', 'date_l': 'date'}
    loserdf.rename(columns = loser_new_names, inplace=True)
    
    df = pd.merge(winnerdf, loserdf, how='inner', on=['winner', 'loser', 'date'])
    
    return df
    
    
```
